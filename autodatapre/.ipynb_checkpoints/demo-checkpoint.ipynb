# Classification Task with a Runtime of 10s

## Step 1. Task Information Import

```python
datasetName = 'datasets/42493.csv'
datasetTarget = 'Delay'
runTime=10
df = pd.read_csv(datasetName, sep=',', encoding='ISO-8859-1')
df
```

## Step 2. Pipeline Generation

```python
detailResult,preparedDataset=AutoDP.Classifier(df,datasetName,datasetTarget,runTime)
```

## Step 3. Statistics and Analysis

```python
AutoDP.EnhancedFunction(df,preparedDataset,detailResult,taskType="CLA")
```

## Step 4. Store the processed dataset locally

```python
AutoDP.store(datasetName,preparedDataset)
```

# Regression Task with Convergence Runtime
