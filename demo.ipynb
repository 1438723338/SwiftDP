```python
import warnings
warnings.filterwarnings("ignore")
import pandas as pd
import sys
import os

current_dir = os.getcwd()
metafeature_path = os.path.join(current_dir, 'autodatapre', 'datasets', 'dataset', 'Metafeature.csv')
label_path = os.path.join(current_dir, 'autodatapre', 'datasets', 'dataset', 'label.csv')

print(sys.version)
```

# **Import our SwiftDP**

```python
import autodatapre as SwiftDP
```

# **Classification Task**

## **Step 1. Task Information Import**

```python
datasetName = os.path.join('autodatapre', 'datasets', '42493.csv')
datasetTarget = 'Delay'
runTime = 10
df = pd.read_csv(datasetName, sep = ',', encoding = 'ISO-8859-1')
df
```

## **Step 2. Pipeline Generation**

```python
detailResult, preparedDataset = SwiftDP.Classifier(df, datasetName, datasetTarget, runTime)
```

## **Step 3. Statistics and Analysis**

```python
SwiftDP.EnhancedFunction(df, preparedDataset, detailResult, taskType = "CLA")
```

## **Step 4. Incremental Update**

```python
SwiftDP.UpdateMetaData(df, detailResult, metafeature_path, label_path)
```
